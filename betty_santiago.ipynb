```python
import math
import numpy as np
from matplotlib import pyplot as plt
import sympy as sp
```

```python
rho = 10

e = 8.854187817*10**(-12)

R = 50
```

```python
a = np.linspace(0, 50, 100000)
x = np.linspace(50, 1000, 100000)
```

```python
y = (rho*R**3)/(3*e*x)
b = ((rho*R**2)/(2*e))-((rho*a**2)/(6*e))
```

```python
def f(x):
    return y
plt.plot(x, f(x))

def h(a):
    return b                         
plt.plot(a, h(a))





plt.ion()
plt.xlabel("r")
plt.ylabel("Phi de r")
plt.title("Representacion de funciones")
plt.plot(x, f(x), color="blue", linewidth=2.5, linestyle="-", label="r > R")
plt.plot(a, h(a), color="red",  linewidth=2.5, linestyle="-", label="r <= R")
plt.legend(loc='upper right')
```

```python
#Declaramos los símbolos que usaremos
x, y = sp.symbols('x y')

#Utilizamos el método init_printing
sp.init_printing(use_unicode=True)
```

```python
d=sp.diff((rho*R**3)/(3*e*x), x)
```

```python
c=sp.diff(((rho*R**2)/(2*e))-((rho*x**2)/(6*e)),x)
```

Recordando que el campo es la derivada, salvo el signo negativo entonces:

```python
print("El campo para r>R es:", -d, "El campo para r <= R es:", -c)
```

Graficando nuevamente

```python
x_1 = np.linspace(5, 50, 10000)
x_2 = np.linspace(50, 1000, 10000)

y_1 = (rho*R**3)/(3*e*x_1**2)
y_2 = (-2*rho*x_2)/(6*e)
```

```python
def j(x_1):
    return y_1
plt.plot(x_1, j(x_1))

def k(x_2):
    return y_2                         
plt.plot(x_2, k(x_2))





plt.ion()
plt.xlabel("r")
plt.ylabel("Phi de r")
plt.title("Representacion de funciones")
plt.plot(x_1, j(x_1), color="yellow", linewidth=2.5, linestyle="-", label="Campo r > R")
plt.plot(x_2, k(x_2), color="black",  linewidth=2.5, linestyle="-", label="Campo r <= R")
plt.legend(loc='upper right')
```
