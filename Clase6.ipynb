### 4.3 Dominio, contradominio y rango de una relación

**Definición.** Sea $R$ una relación de $X$ en $Y$, entonces


(a) decimos que el **domininio** de $R$, $Dom(R)$, es el conjunto

$$
Dom(R)= \{x \in X| \exists y\in Y \quad \text{tal que} \quad (x,y)\in R \}
$$


(b) además, la **imagen o rango** de $R$, $Im(R)$, es el conjunto


$$
Im(R)=\{ y\in Y| \exists x\in X \quad \text{tal que} \quad (x,y)\in R  \}
$$

$Im(R)=\{ \}$

**Definición.** Sean $R$ una relación de $X$ en $Y$ y $A\subseteq X$, $B\subseteq Y$. La imagen de $A$ bajo $R$, $R[A]$, y la imagen inversa de B bajo $R$,$R^{-1}[B]$, están dadas, respectivamente por


$$
    R[A]=\{y\in Y| \exists x\in A \quad \text{tal que} \quad (x,y)\in R \}
$$


$$
    R^{-1}[B]=\{x\in X| \exists y\in B \quad \text{tal que} \quad (x,y)\in R \}
$$




La relación $R^{-1}$ se llama la **relación inversa** de $Y$ en $X$ y está dada por

$$
R^{-1}= \{(y,x)|(x,y)\in R\}.
$$
