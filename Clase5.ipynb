### 4.1 Relaciones entre dos conjuntos

#### Par ordenado

Con el objetivo de tener un orden o diferenciación en las parejas de conjuntos relacionados, es necesario definir a un par ordenado $(\alpha,\beta)$ de modo que si existe otro par ordenado $(\gamma,\eta)$, tengamos que son iguales únicamente si


$$
\alpha=\gamma \quad \text{y} \quad \beta=\eta.                                                          (1)
$$

La definición que cumple con este propósito es la siguiente:


**Definición.** Sean dos conjuntos $A$ y $B$ no vacíos y $\alpha \in A$, $\beta \in B$. Decimos que el *par ordenado* de $\alpha$ y $\beta$ está dado por el conjunto


$$
(\alpha,\beta) \equiv \{ \alpha, \{\alpha,\beta \}  \}. 
$$


Notemos con los siguientes ejemplos que efectivamente se cumple (1).



**Ejemplos.** 



1.- Sean los pares ordenados (1,2) y (2,1). Vemos que, de acuerdo con la definición anterior,

$$
(1,2) = \{1,\{1,2\} \}
$$

mientras que


$$
(2,1) = \{2,\{2,1\} \}= \{2,\{1,2\} \}.
$$


Por lo que, $(1,2) \neq (2,1)$. 

En el caso de los pares ordenados (3,4) y (tres,cuatro), éstos son

$$
(3,4) = \{3,\{3,4\} \}
$$

y

$$
(tres,cuatro) = \{tres,\{tres,cuatro\} \},
$$

es decir $(3,4) \neq (tres,cuatro)$.

Un ejemplo donde sí se de la igualdad es el caso de (0,2) y $(a,b)=\{a\in \text{\{neutros aditivos\}}\quad \text{y} \quad  b\in \{\text{m.c.d de 6,8,12,15}\}  \}$.


El neutro aditivo es un número $y$ tal que $\forall x\in \mathbb{R}$ x+y=x. Esa $y$ es el *neutro aditivo* de los números reales y es único, tiene el símbolo $y \equiv 0$.


 $a=0 \quad \text{y}\quad \{b\in \{\text{m.c.d de 6,8,12}\}  \}$.

Entonces $b=2$.

(a,b)=(0,2). Entonces, es claro que (0,2)=(0,2) con la definición de par ordenado.




Ahora bien, ya sabemos qué es una pareja ordenada pero no sabemos cómo formarla. Es decir, mediante las operaciones entre conjuntos definidas anteriormente, ¿cómo logramos obtener un par ordenado?. 



Es imposible hacer un par ordenado con unión, intersección, resta o complemento. Es necesario definir una nueva operación: el producto cartesiano.


#### Producto cartesiano

**Definición.** Sean $A$ y $B$ dos conjuntos no vacíos. Definimos el conjunto *producto cartesiano* $A\times B$, como


$$A \times B \equiv  \{ (a,b)| a\in A \quad \text{y} \quad b\in B\}.$$


Es decir, son todos los pares ordenados (a,b) tales que se cumple $a\in A$ **y** $b \in B$.



**Ejemplos.**


Sean $A=\{1,2,a \} $ y $B=\{0,3\}$. El producto cartesiano de estos conjuntos es 


$$
A\times B= \{ (1,0),(1,3),(2,0),(2,3),(a,0),(a,3)\},
$$

notemos que


$$
B\times A= \{ (0,1),(0,2),(0,a),(3,1),(3,2),(3,a)\}.
$$

En general tenemos que $A\times B \neq B\times A$, a menos que $A=B$. 


$$
A \times A= \{(1,1),(1,2),(1,a),(2,1),(2,2),(2,a),(a,1),(a,2),(a,a)\},
$$

mientras que




$$
B\times B= \{(0,0),(0,3),(3,3),(3,0)\}.
$$

## Relaciones binarias

A continuación agregaremos un ingrediente más a nuestra receta para llegar al cálculo moderno. Necesitamos un conjunto de reglas o instrucciones que nos permitan fabricar conjuntos de pares ordenados. Este conjunto de instrucciones se llama una *relación*, se denota por la letra $R$ y existe una gran variedad de ellas. Consideremos el conjunto $\mathcal{R}$ formado por todas las relaciones $R$ entre dos conjuntos no vacíos, es decir $\mathcal{R}=\{R|R \hspace{1mm}\text{ es una relación binaria} \}$.


**Definición.** Sean X y Y conjuntos no vacíos. Una *relación de X en Y*, es una pareja ordenada $(R, X\times Y)$, donde $R\subseteq X\times Y$. Es decir, R es un conjunto formado de parejas ordenadas que son elementos de $X \times Y$ (observación: no todos los elementos de R son elementos de $X\times Y$). Si $R$ es una relación de $X$ en $Y$ y $(x,y) \in R$, decimos que $x$ está relacionado con $y$ mediante $R$, o bien $xRy$.

Para hacer un par ordenado necesitamos entonces


* Dos conjuntos no vacíos X, Y.

+ Una relación binaria de X en Y o de Y en X, R.


Hagamos unos **ejemplos:**

$R=\phi$.


1.1) Si consideramos el caso de tomar como $X=A$ y $Y=A$, es decir $(R, A\times A)$, en ese caso

$R_{1}= \{(1,1),(2,2),(a,a) \}$.


1.2) Si ahora consideramos el caso de tomar como $X=B$ y $Y=B$, es decir $(R, B\times B)$, en ese caso

$R= \{ (1,0),(1,3),(2,0),(2,3),(a,0),(a,3)\}$. En este caso **particular**, $R = A\times B$.



1.4) Si consideramos el caso de tomar como $X=A$ y $Y=A$, es decir $(R, A\times A)$, en ese caso

$R_{2}=\{(1,2),(1,a),(2,1),(2,a),(a,1),(a,2)\}$.


Observación.

Para el siguiente ejemplo, consideremos un conjunto cualquiera $X$. 

**Definimos** el *conjunto potencia*, $\mathcal{P}(X)$, como el conjunto formado por todos los subconjuntos de $X$. Es decir,

$$
\mathcal{P}(X)\equiv  \{A|A\subseteq X \}.
$$

Por ejemplo, si $X=\{1,2,3\}$ entonces $\mathcal{P}(X)= \{\{1\},\{2\},\{3\},\{1,2\},\{1,3\},\{2,3\},\{ 1,2,3\} \}$.

En el caso particular de que $X=\phi$, $\mathcal{P}(X)= \{ \phi \}$.

*Observación.* La cardinalidad del conjunto potencia de un conjunto $A$ cualquiera, es $\#(\mathcal{P}(A))= 2^{\#(A)}$. Para demostrar esto, basta contar la cantidad de formas en las que podemos elegir $k$ elementos de un total de $n$, con $0\leq k \leq n$, donde $n=\#(A)$.

${\displaystyle |{\mathcal {P}}(A)|={n \choose 0}+{n \choose 1}+\ldots +{n \choose k}+\ldots +{n \choose n}=2^{n}=2^{|A|}}$


Para un número $k$ cualquiera, esta cantidad es ${n \choose k}$, lo cual denota las *combinaciones de $n$ en $k$* y está dada por

$${n \choose k}= \frac{n!}{k!(n-k)!}.$$

En donde $n!$ es el *factorial de n*, y se define como $$n!\equiv n(n-1)(n-2)....(2)(1),$$
 
para cualquier número natural $n$. Cuando $n=0$, se toma por convención que $n!=0!\equiv 1$.
 


4!=4*3*2*1=24

5!=5*4*3*2*1=5*4!=120

6!=6*5!=720

7!=7*6!=5040
