### Operaciones con conectores lógicos

Como se mencionó en la clase anterior, la lógica proposicional o aristotélica es útil para evaluar proposiciones mediante operadores y posteriormente, obtener el valor de verdad de estas operaciones.

**Tautología y contradicción**

Una proposición es una *tautología* si su valor de verdad siempre es verdadero sin importar el valor de verdad de las proposiciones que la componen. 
Particularmente, en el caso de que la proposición $P \leftrightarrow Q$ sea una tautología, significa que P y Q son tautológicamente equivalentes, es decir, P es verdadera si y sólo si Q lo es. De modo que, si este es el caso y demostrar Q es más fácil que demostrar P, podemos hacerlo ya que P será verdadera.

## Métodos de demostración

En base a lo anterior, a continuación se mencionan los métodos de demostración más utilizados:


**a) Demostración directa:**
Se parte de la hipótesis $P$ y se obtiene $Q$ mediante argumentos lógicos, definiciones y teoremas previamente demostrados.

**b) Demostración por contraposición:**
Se utiliza la definición de tautología con la proposición $(P \rightarrow Q)$ y su equivalencia con $ [(¬Q)\rightarrow (¬P)]$.

**Ejemplo.**
Consideremos las proposiciones dadas por

*P: $x$ es impar*

*Q: $x+2$ es impar*


Demostraremos que $P\rightarrow Q$ es verdadera mediante la demostración de que $¬Q \rightarrow ¬P$ es verdadera. La negación de estos dos enunciados es:

*¬P: $x$ es par*

*¬Q: $x+2$ es par*

Por lo tanto, ¬Q significa que existe un número $k$ tal que $x+2=2k$. 
Restando 2 de ambos lados obtenemos: $x=2k-2$,
factorizando el 2: $x=2(k-1)$. Nombrando a $k-1\equiv m$, obtenemos que $x=2m$. Es decir, $x$ es un número par. 

Hemos obtenido a partir de tomar ¬Q como verdadera, que ¬P es verdadera, por lo tanto la implicación $¬Q \rightarrow ¬P$ es verdaera, y como es una tautología de $ P \rightarrow Q$, entonces ésta también es verdadera.


**c) Demostración por reducción al absurdo:**

Para demostrar la proposición $ P \rightarrow Q$, podemos negar la conclusión, es decir tomar $¬Q$ y P como verdaderas y llegar a una contracción del tipo R $\wedge$ $¬R$.

**Ejemplo.**

Demostremos (por reducción al absurdo) que para cualquier par de números reales $x$,$y$ $xy=0$ implica que alguno de los dos es cero, es decir que $x=0$ o $y=0$. 

Para comenzar, supongamos que $xy=0$ es verdadera y que la implicación no lo es, es decir que $x\neq0$ y $y\neq0$, 
ahora tratemos de llegar a una contradicción. 

Para eso notemos que cualquier número entero $z$ cumple que $z*=0$. Por lo que en particular,

$$xy=x*0=0$$

pero como tenemos que $x\neq 0$ (por hipótesis), entonces de la primera parte de la igualdad anterior, obtenemos que $y=0$, lo cual es una contradicción, pues habíamos supuesto que $y\neq 0$.
